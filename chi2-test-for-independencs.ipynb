```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/kaggle/input/chi-square-tests/students.csv")
data.head()
```

```python
cdf = data[["gender", "religion"]]
cdf.head()
```

chi-square test to see if gender and relation has a relation with each other in our date

```python
contingency = pd.crosstab(cdf['gender'], cdf['religion'])
'''print(contingency)
print(type(contingency))'''

contingency['total_gender'] = contingency['Catholic'] + contingency['Muslim'] + contingency['Orthodox'] + contingency['Other'] + contingency['Protestant']
total_religion = {'Catholic' : [2797], 'Muslim': [330], 'Orthodox': [585], 'Other' : [2688], 'Protestant': [1839], 'total_gender':[8239]}
total_religion = pd.DataFrame(total_religion)
contingency = pd.concat([contingency, total_religion])
cintingency = pd.DataFrame(contingency, index=['Female', 'Male', 'total_religion'])
contingency
```

```python
 expected_table = np.array([[
    (contingency['Catholic'].sum() / 2) *(contingency.loc['Female'].sum() / 2) / contingency.iloc[2][5],
    (contingency['Muslim'].sum() / 2) * (contingency.loc['Female'].sum() / 2) / contingency.iloc[2][5],
    (contingency['Orthodox'].sum() / 2) * (contingency.loc['Female'].sum() / 2) / contingency.iloc[2][5],
    (contingency['Other'].sum() / 2) * (contingency.loc['Female'].sum() / 2) / contingency.iloc[2][5],
    (contingency['Protestant'].sum() / 2) *(contingency.loc['Female'].sum() / 2) / contingency.iloc[2][5]],
    [
    (contingency['Catholic'].sum() / 2) * (contingency.loc['Male'].sum() / 2) / contingency.iloc[2][5],
    (contingency['Muslim'].sum() / 2) * (contingency.loc['Male'].sum() / 2) / contingency.iloc[2][5],
    (contingency['Orthodox'].sum() / 2) * (contingency.loc['Male'].sum() / 2) / contingency.iloc[2][5],
    (contingency['Other'].sum() / 2) * (contingency.loc['Male'].sum() / 2) / contingency.iloc[2][5],
    (contingency['Protestant'].sum() / 2) * (contingency.loc['Male'].sum() / 2) / contingency.iloc[2][5],
    ]])
    
print(expected_table)
```

```python
chi_2 = 0
for i in range(2):
    for j in range(5):
        chi_2 += ((contingency.iloc[i][j] - expected_table[i][j]) ** 2) / expected_table[i][j]
print(chi_2)

num_rows , num_columns = expected_table.shape
degrees_of_freedom = (num_rows - 1) * (num_columns - 1)
print(degrees_of_freedom)
p_value = 0.17129
```

```python
alpha = 0.05
# according to distribution of critical value:
# alpha = 0.05 , degrees of freedom = 4  : critical value = 9.49

if chi_2 > 9.49:
    print('H0 rejected, there is a relation between sex and religion')
else:
    print('H0 accepted, , there is not a relation between sex and religion')
    
if p_value <= alpha:
    print('H0 rejected, there is a relation between sex and religion')
else:
    print('H0 accepted, , there is not a relation between sex and religion')
```
