```python
import pandas as pd
```

```python
data = pd.read_csv("/kaggle/input/chi-square-tests/students.csv")
data.head()
```

```python
cdf = data[["gender", "religion"]]
cdf.head()
```

chi-square test to see if gender and relation has a relation with each other in our date

```python
contingency = pd.crosstab(cdf['gender'], cdf['religion'])
'''print(contingency)
print(type(contingency))'''

contingency['total_gender'] = contingency['Catholic'] + contingency['Orthodox'] + contingency['Other'] + contingency['Protestant']
total_religion = {'Catholic' : [2797], 'Muslim': [330], 'Orthodox': [585], 'Other' : [2688], 'Protestant': [1839], 'total_gender':[7909]}
total_religion = pd.DataFrame(total_religion)
contingency = pd.concat([contingency, total_religion], ignore_index = True)
print(contingency)

'''
0 is representing females
1 is representing males
and 2 is representing total religion
'''
```
